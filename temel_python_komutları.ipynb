### 🔸 Soru 10:
`while` döngüsü ile kullanıcının girdiği sayıya (n=5) kadar Fibonacci dizisini yazdırın.

Örnek çıktı: `0, 1, 1, 2, 3`

```python
# Cevabınızı buraya yazın:
n = 5
a, b = 0, 1
sayac = 0

print(f"{n} terimli Fibonacci dizisi:")

while sayac < n:
    print(a, end=", " if sayac < n - 1 else "")

    yeni_sayi = a + b
    a = b
    b = yeni_sayi
    sayac += 1
```

### 🔸 Soru 15:
Bir sayının asal olup olmadığını kontrol eden `asal_mi(n)` fonksiyonu yazın.
- 1'den 20'ye kadar olan asal sayıları bu fonksiyonu kullanarak yazdırın.

```python
# Cevabınızı buraya yazın:
def asal_mi(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

asal_sayilar = []
for sayi in range(1, 21):
    if asal_mi(sayi):
        asal_sayilar.append(sayi)

print(f"1-20 arasındaki asal sayılar: {asal_sayilar}")
```

### 🔸 Soru 16:
Bir listenin ortalamasını, en büyük ve en küçük elemanını döndüren `liste_istatistik(liste)` fonksiyonu yazın.
Fonksiyon 3 değer döndürsün: `(ortalama, minimum, maksimum)`

```python
# Cevabınızı buraya yazın:
def liste_istatistik(liste):
    if not liste:
        return 0, 0, 0

    ortalama = sum(liste) / len(liste)
    minimum = min(liste)
    maksimum = max(liste)

    return ortalama, minimum, maksimum

sayilar = [10, 20, 30, 40, 50]
ort, en_kucuk, en_buyuk = liste_istatistik(sayilar)

print(f"Liste: {sayilar}")
print(f"Ortalama: {ort}")
print(f"En Küçük: {en_kucuk}")
print(f"En Büyük: {en_buyuk}")
```

### 🔸 Soru 17:
Kullanıcıdan bir sayı alındığını varsayın (`girdi = "abc"`).
Bu değeri int'e çevirmeye çalışırken `ValueError` hatası oluşabilir.
try-except ile bunu yakalayan ve anlamlı bir mesaj veren kod yazın.

```python
# Cevabınızı buraya yazın:
girdi = "abc"

try:
    sayi = int(girdi)
    print(f"Başarılı! Girdiğiniz sayı: {sayi}")

except ValueError:
    print("Hata: Geçersiz giriş! Lütfen sadece rakamlardan oluşan bir tam sayı girin.")

finally:
    print("İşlem tamamlandı.")
```

### 🔸 Soru 18 (Bütünleştirici):
Bir **not defteri uygulaması** yazın:
- `notlar` adında boş bir liste oluşturun
- `not_ekle(metin)` fonksiyonu: Notu listeye ekler
- `not_listele()` fonksiyonu: Tüm notları numaralandırarak yazdırır
- `not_sil(index)` fonksiyonu: Verilen indexteki notu siler, hatalı index için hata mesajı verir

Fonksiyonları test edin: 3 not ekleyin, listeleyin, ortadakini silin, tekrar listeleyin.

```python
# Cevabınızı buraya yazın:

notlar = []

def not_ekle(metin):
    notlar.append(metin)
    print(f"Not eklendi: '{metin}'")

def not_listele():
    print("\n--- Güncel Not Listesi ---")
    if not notlar:
        print("Not defteri şu an boş.")
    else:
        for sira, not_metni in enumerate(notlar, start=1):
            print(f"{sira}. {not_metni}")
    print("--------------------------")

def not_sil(index):
    try:
        silinen = notlar.pop(index - 1)
        print(f"Başarıyla silindi: '{silinen}'")
    except IndexError:
        print(f"Hata: {index} numaralı bir not bulunamadı!")
    except Exception as e:
        print(f"Beklenmedik bir hata oluştu: {e}")


not_ekle("Marketten süt al")
not_ekle("Python ödevini tamamla")
not_ekle("Akşam saat 20:00'de toplantı var")

not_listele()

not_sil(2)

not_listele()
```
