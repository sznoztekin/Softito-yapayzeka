### 🔸 Soru 12:
Bir öğrenci listesi var. Her öğrenci `(isim, not)` tuple'ı olarak saklanıyor.
- Notu 70'in üzerinde olanları filtreleyin
- En yüksek notu alan öğrenciyi bulun
- Notların ortalamasını hesaplayın

```python
# Cevabınızı buraya yazın:
ogrenciler = [
    ("Ahmet", 85),
    ("Büşra", 62),
    ("Can", 91),
    ("Deniz", 74),
    ("Elif", 58),
    ("Furkan", 88)
]

basarili=[(isim,not_) for isim,not_ in ogrenciler if not_ >70]
print("70 üzeri öğrenciler: ")
for isim,not_ in basarili:
    print(f"{isim} : {not_}")

en_iyi = max(ogrenciler, key=lambda x: x[1])
print(f"\nEn yüksek not: {en_iyi[0]} → {en_iyi[1]}")

ortalama = sum(not_ for _, not_ in ogrenciler) / len(ogrenciler)
print(f"Notların ortalaması: {ortalama:.1f}")
```

### 🔸 Soru 14:

### 🔸 Soru 15:
Özyinelemeli (recursive) bir fonksiyon yazın: `faktoriyel(n)` - n! hesaplasın.
- 0! = 1
- n! = n × (n-1)!

5!, 10! değerlerini hesaplayın.

```python
# Cevabınızı buraya yazın:
def faktoriyel(n):
    if n==0:
      return 1
    return n* faktoriyel(n-1)
print(faktoriyel(5))
print(faktoriyel(10))
```

### 🔸 Soru 18 (Bütünleştirici):
Basit bir **öğrenci not sistemi** yazın:

- `ogrenciler` adında boş bir sözlük oluşturun (anahtar: isim, değer: notlar listesi)
- `ogrenci_ekle(isim)`: Yeni öğrenci ekler
- `not_ekle(isim, not_)`: Öğrenciye not ekler. Öğrenci yoksa hata mesajı verir.
- `ortalama_hesapla(isim)`: Öğrencinin not ortalamasını döndürür
- `sinif_raporu()`: Tüm öğrencileri ve ortalamalarını yazdırır, sınıf ortalamasını gösterir

En az 3 öğrenci ve her birine 3 not girerek test edin.

```python
# Cevabınızı buraya yazın:
ogrenciler = {}

def ogrenci_ekle(isim):
    if isim in ogrenciler:
        print(f"{isim} zaten kayıtlı!")
    else:
        ogrenciler[isim] = []
        print(f"{isim} eklendi.")

def not_ekle(isim, not_):
    if isim not in ogrenciler:
        print(f"Hata: {isim} bulunamadı!")
    else:
        ogrenciler[isim].append(not_)
        print(f"{isim} → {not_} notu eklendi.")

def ortalama_hesapla(isim):
    if isim not in ogrenciler or len(ogrenciler[isim]) == 0:
        return 0
    return sum(ogrenciler[isim]) / len(ogrenciler[isim])

def sinif_raporu():
    print("\n📊 Sınıf Raporu")
    print("-" * 30)
    for isim in ogrenciler:
        ort = ortalama_hesapla(isim)
        print(f"{isim:10} → Ortalama: {ort:.1f}")
    print("-" * 30)
    tum_ortalamalar = [ortalama_hesapla(i) for i in ogrenciler]
    sinif_ort = sum(tum_ortalamalar) / len(tum_ortalamalar)
    print(f"Sınıf Ortalaması: {sinif_ort:.1f}")

# Test
ogrenci_ekle("Ali")
ogrenci_ekle("Ayşe")
ogrenci_ekle("Can")

not_ekle("Ali", 85)
not_ekle("Ali", 90)
not_ekle("Ali", 78)

not_ekle("Ayşe", 92)
not_ekle("Ayşe", 88)
not_ekle("Ayşe", 95)

not_ekle("Can", 70)
not_ekle("Can", 65)
not_ekle("Can", 72)

sinif_raporu()

# Test kodu:
# ogrenci_ekle("Ahmet")
# ogrenci_ekle("Büşra")
# ogrenci_ekle("Can")
# not_ekle("Ahmet", 85)
# not_ekle("Ahmet", 90)
# not_ekle("Ahmet", 78)
# not_ekle("Büşra", 92)
# not_ekle("Büşra", 88)
# not_ekle("Büşra", 95)
# not_ekle("Can", 70)
# not_ekle("Can", 65)
# not_ekle("Can", 73)
# sinif_raporu()
```
